```python
df['combined_feature']=df['keywords']+df['cast']+df['genres']+df['director']
df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['combined_feature'])
```

```python
print(vectorizer.get_feature_names())
```

```python
print(X.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs=cosine_similarity(X.toarray())
```

```python
print(cs)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]
def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
movie_user_likes = "Avatar"
movie_index = get_index_from_title(movie_user_likes)
similar_movies = list(enumerate(cs[movie_index])) 
#accessing the row corresponding to given movie to find all the similarity scores for that movie and then enumerating over it
```

```python
similar_movies
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
print("Top 5 similar movies to "+ movie_user_likes +" are:\n")
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>5:
        break
```
